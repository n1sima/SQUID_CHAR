```python
def  newplotVPhi0(fname):
    
    nhead = 12
    toVolt = 39.3216/((2**16) -1)
    fd = open(fname, 'rb')
    read_data = np.fromfile(file=fd, dtype=np.int16)#.reshape(shape)
    fd.close()
    dat = toVolt*read_data[nhead:]
    print(dat)
    xs = np.arange(len(dat))
    y=dat**2
    min_index = np.argmin(y)
    min_x = xs[min_index]
    min_y = y[min_index]
    Nav=min_x-300000
    V0 = np.mean(dat[:Nav])
    V2 = np.mean(dat[-Nav:])
    V=1e3*(V0-V2)
    plt.plot(xs, dat)
    print(V0,V2,V)
    plt.plot(xs, dat)
    plt.axhline(y=V0, color='r', linestyle='--', label='y = V0 ({:.2f})'.format(V0))
    plt.axhline(y=V2, color='g', linestyle='--', label='y = V2 ({:.2f})'.format(V2))
    plt.axvline(x=Nav, color='r', linestyle='--', label=r'$Nav = {:.2f}$'.format(Nav))
    plt.axvline(x=len(dat)-Nav, color='g', linestyle='--', label=r'$Len(data)-Nav = {:.2f}$'.format(len(dat)-Nav))
    #plt.title(fname)
    #plt.legend()
    plt.grid(True)
    plt.xlabel(r'ms', fontsize= 16)
    plt.ylabel(r'V [A]', fontsize= 16)
    downloads_path = os.path.expanduser(r"C:\Users\n1cos\OneDrive\Documentos\Semester 6\Bachelor Thesis")
    file_path = os.path.join(downloads_path, "SQUIDResets.jpg")
    plt.savefig(file_path)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

RValues = [1,1.5]
x = np.linspace(-5.5, 5.5, 1000)

plt.figure(figsize=(8, 6))

for i, R in enumerate(RValues):
    f = lambda x: np.sin(2*np.pi * x) + np.abs(np.sin(2*np.pi * R * x))
    plt.plot(x, f(x), label=f'R = {R}')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plots von f(x) fr verschiedene R-Werte')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a time array with a smaller interval for smoother plotting
t = np.linspace(-0.5, 6, 5000)  # Extend the time interval for smoother curve
sinus_wave = np.sin(2 * np.pi * t)

# Define the working points as before
working_points = []
for i in range(1, len(t) - 1):
    if sinus_wave[i] > 0 and sinus_wave[i - 1] <= 0:
        working_points.append((t[i], 0))

# Create horizontal arrows to the right at specific heights and lengths
arrows = [
    # Arrow at 1 V, 2 periods long (red)
    (1.0, 1, 3.7, 1, 'red'),
]

# Add a small arrow going to the left, 0.3 periods long (blue)
arrows.append((3.7, 0, 4.0, 0, 'blue'))

# Plot the sinus wave, working points, and arrows
plt.plot(t, sinus_wave)
for x, y in working_points:
    plt.scatter(x, y, color='red', marker='o',s=150)

for start_x, start_y, end_x, end_y, color in arrows:
    plt.annotate('', xy=(end_x, end_y), xytext=(start_x, start_y),
                 arrowprops=dict(arrowstyle='->', color=color,linewidth=3))

# Set the x-axis, y-axis labels, and axis limits
plt.xlabel(r'$ \frac{\phi_a}{\phi_0}$', fontsize=18)
plt.ylabel('V', fontsize=18)
plt.xlim(-0.5, 5)
plt.grid()
downloads_path = os.path.expanduser(r"C:\Users\n1cos\OneDrive\Documentos\Semester 6\Bachelor Thesis")
file_path = os.path.join(downloads_path, "Resetfunction.jpg")
plt.savefig(file_path)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os  # Import the os module

# Erzeugen der Sinuswelle im Intervall [0, 6]
t = np.linspace(-0.5, 3, 1000)
sinus_wave = np.sin(2 * np.pi * t)

# Arbeitspunkte (W) finden
working_points = []
for i in range(1, len(t) - 1):
    if sinus_wave[i] > 0 and sinus_wave[i - 1] <= 0:
        working_points.append((t[i], 0))

# Erzeugen der horizontalen Pfeile
arrows = []
for x, y in working_points:
    left_arrow = (x - 0.5, x)  # Pfeil nach links
    right_arrow = (x, x + 0.5)  # Pfeil nach rechts
    arrows.extend([left_arrow, right_arrow])

# Plotten der Sinuswelle, Arbeitspunkte und Pfeilen
plt.plot(t, sinus_wave)
for x, y in working_points:
    plt.scatter(x, y, color='red', marker='o', s=150)  # Adjusted marker size using 's'

# Abwechselnd Pfeile nach links und rechts zeichnen
for i in range(0, len(arrows), 2):
    start, end = arrows[i]
    plt.annotate('', xy=(end, 0), xytext=(start, 0),
                 arrowprops=dict(arrowstyle='->', color='red', linewidth=3))  # Removed space in arrowstyle
for i in range(1, len(arrows), 2):
    start, end = arrows[i]
    plt.annotate('', xy=(end, 0), xytext=(start, 0),
                 arrowprops=dict(arrowstyle='<-', color='blue', linewidth=3))  # Removed space in arrowstyle

plt.xlabel(r'$ \frac{\phi_a}{\phi_0}$', fontsize=18)
plt.ylabel('V', fontsize=18)
plt.xlim(-0.5, 1.55)
plt.grid()

downloads_path = os.path.expanduser(r"C:\Users\n1cos\OneDrive\Documentos\Semester 6\Bachelor Thesis")
file_path = os.path.join(downloads_path, "Resetexplain.jpg")
plt.savefig(file_path)
plt.show()

```
